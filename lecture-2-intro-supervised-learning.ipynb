## Supervised machine learning paradigm
Usual programming approach:
- Start with a specification of the desired behavior.
- Think about logic that will achieve that behavior.
- Write program

The ML approach
- Examples of the desired behavior (instead of a formal specification)
- ML Algorithm (code we write)
- Produced "program" (model)

## Notation:
- "Input": the input that we will provide to the ML program
- "Desired outputs": the output we want our ML program for some corresponding input.
- "Training Set": a collection of inputs and their corresponding desired outputs.

## The three ingredients of ML algorithm
1. Model (i.e "ML program"):
  - function that maps inputs to predicted outputs. `h: Inputs -> Predicted Outputs`.
  - Model class / architecture: family of models we consider.
2. Loss function
  - function that measures the difference between predicted and desired outputs. `L: Predicted Output x Desired Output -> Real number`
  - loss is "large" if prediction is bad and "small"/zero if prediction is good.
3. Optimization procedure (training)
  - find the model (or a model) within the model class that achieves low loss on the training set.
  - low sum of loss over all predicted and desired outputs on the training set.

## Generalization
- ML finds a "good" model on the training set.
- But we want a model that performs well on new examples.
- So we use a **test set**: and additional set of examples (inputs and desired outputs) that we do not use to train (optimize) the model.
- For "general" AI: test on "downstream" tasks i.e how it works on actual problems e.g solving a math problem using a large language model instead of just predicting the next word.
