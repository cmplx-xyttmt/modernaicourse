## Vectors
- Vectors are 1-D arrays of numbers. $x \in \mathbb{R} ^ n$
- e.g $x \in \mathbb{R} ^ n = \begin{bmatrix} 4 \\ 3 \\ 7 \end{bmatrix}$
- $x_i$ refers to the ith element of x
- Use one-indexing for vectors (in math)
- Operations on vectors:
  - Addition
  - Scalar multiplication
  - Vector-vector multiplication (inner product)
    - Given $x, y \in \mathbb{R}$, then $<x, y> = x^T y = x_1 y_1 + ... + x_n y_n =\sum_{i = 1}^{n} x_i y_i$


```python
import torch

x = torch.tensor([1.,2.,3.])
y = torch.tensor([4.,5.,6.])
print(f"x: {x}. y: {y}")
print(f"x + y: {x + y}")
print(f"5 * x = {5 * x}")
print(f"inner_product(x, y) = {x @ y}")
```

## Matrices
- A matrix is a 2D array of numbers. $A \in \mathbb{R}^{m \times n}$
- e.g $$A \in \mathbb{R}^{2 \times 3} = \begin{bmatrix} 7 & 3 & 2 \\ 2 & 4 & 8 \end{bmatrix}$$
- Refer to entry in the ith row and jth column as $A_{i j}$
- Operations:
  - Matrix addition.
  - Matrix Scalar multiplication.
  - Transpose: $A^T$. Swap rows and columns i.e $(A^T)_{i j} = A_{j i}$

```python
A = torch.tensor([[7., 3, 2], [2, 4, 8]])
print(f"A[0][1] = {A[0,1]}")
print(f"first row of A: {A[0,:]}")
print(f"third column of A: {A[:,2]}")
print(f"transpose of A: {A.T}")
```

## Matrix-vector products
- Definition: $A \in \mathbb{R}^{m \times n}, x \in \mathbb{R}^n$ then $A x \in \mathbb{R}^m$ where $(A x)_i = \sum_{j = 1}^n {A_{i j} x_j}$
- Interpretation #1: collection of inner products.
- Interpretation #2: linear combination of the columns of A with coefficients given by entries of x.
- For $x \in \mathbb{R}^m, A \in \mathbb{R}^{m \times n}$ then $x^T A \in \mathbb{R}^{1 \times m}$ where $(x^T A)_i = \sum_{j = 1}^n {x_j A_{j i}}$
  - Interpretation #1: collection of inner products.
  - Interpretation #2: linear combination of the rows of A.

```python
A = torch.tensor([[7., 3, 2], [2, 4, 8]])
B = torch.tensor([[4., 2, 3], [1, 8, 9]])
x = torch.tensor([1.,2.,3.])
y = torch.tensor([1., 2])
print(f"Ax = {A@x}")
print(f"Ax collection of inner products interpretation: {A[0,:] @ x},{A[1,:] @ x}")
print(f"Ax linear combination of columns interpretation: {x[0] * A[:,0] + x[1] * A[:,1] + x[2] * A[:,2]}")
print(f"y shape: {y.shape}\ny transpose shape: {y.T.shape}")
print(f"transpose(y) A = {y.T @ A}")
print(f"yA shape: {(y.T @ A).shape}")
print(f"yA inner products: {y.T @ A[:,0]}, {y.T @ A[:,1]}, {y.T @ A[:,2]}")
print(f"yA linear combination of rows: {y[0] * A[0,:] + y[1] * A[1,:]} ")
```

## Matrix-Matrix multiplication
- Given $A \in \mathbb{R}^{m \times n}, B \in \mathbb{R}^{n \times p}$, then $A B \in \mathbb{R}^{m \times p}$ where $(A B)_{i j} = \sum_{k = 1}^{n} A_{i k} B_{k j}$
- Interpretation #1: inner products of all rows of $A$ with columns of $B$.
- Interpretation #2: collection of matrix-vector products i.e product of matrix A with the columns of B.
- Interpretation #3: collection of vector-matrix products i.e product of the rows of matrix A with B.
- Properties of matrix multiplication:
  - Distributive: $A (B + C) = A B + A C$
  - Associative: $(A B) C = A (B C)$
  - Not commutative: $A B \neq B A$
  - Transpose of product: $(A B) ^ T = B^T A^T$

```python
A = torch.tensor([[7., 3, 2], [2, 4, 8]])
B = torch.tensor([[5., 6, 7], [4, 2, 3], [3, 2, 5]])
print(f"AB = {A @ B}")
print(A.shape, B.shape)
```

```python
# can have 3 or more dimensional tensors
A = torch.tensor([[[7., 3, 2], [2, 4, 8]], [[5., 6, 7], [4, 2, 3]]])
print(f"A shape: {A.shape}")
print(A[1,:,:])
```
